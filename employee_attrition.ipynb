```python
data_clean = dataset.drop(columns=['EmployeeCount', 'Over18', 'StandardHours', 'EmployeeNumber'])
```

```python
from sklearn.preprocessing import LabelEncoder                                  #LabelEncoder used to convert categorical labels into numerical labels
label_encoders = {}
for column in data_clean.select_dtypes(include=['object']).columns:
  le = LabelEncoder()
  data_clean[column] = le.fit_transform(data_clean[column])         #fit method identifies the unique categories in the data. #transform method converts the categories into integers
  label_encoders[column] = le
```

```python
sns.countplot(data=dataset, x='Attrition', palette = 'Set2')                    #palette - used to define the color scheme of the plot
plt.title('Attrition Distribution')
plt.show()

#this graph shows count of employees("NO" bar means employees who stayed means most of the employees did not leave the company)
                                    #("YES" bar means employees who left)
```

```python
plt.figure(figsize=(7, 5))
sns.boxplot(x='Attrition', y='Age', data=data_clean, palette='Set1')
plt.title('Age vs Attrition')
plt.show()
```

```python
sns.countplot(data=data_clean, x='Department', hue='Attrition',palette='Set1')
plt.title("Attrition by Department")
plt.show()
```

0 means - employees who did not leave the company(No Attrition)

1 means - employees who left the company(Yes Attrition)

This graph shows:-

e.g If the "Research & Develpment"(1)department has a relatively high number of employees leaving (Attrition = 1), it might indicate:

Job dissatisfaction.
Stressful work conditions.
A need for intervention to retain employees.

If the "HR"(2) department shows minimal attrition (Attrition = 1), it suggests higher stability in that department.

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

x= data_clean.drop(columns=['Attrition'])              #x: This is the feature matrix, which contains all columns except the target variable Attrition.
y = data_clean['Attrition']                            #y: This is the target variable, which indicates whether an employee left the company (1) or stayed (0).

#train-test-split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2, random_state=42)
#train randomforestclassifier
model= RandomForestClassifier(random_state=42,n_estimators=100)
model.fit(x_train, y_train)
#predictions
y_pred= model.predict(x_test)

print("Classification report:\n", classification_report(y_test, y_pred))
print("Confusion_matrix:\n",confusion_matrix(y_test, y_pred))
importances = model.feature_importances_
features = x.columns
sns.barplot(x=importances, y=features)
plt.title("Feature Importance")
plt.show()
```

```python
#RandomForestClassifier - combines multiple individual decision trees to create a more powerful predictive model
#classification_report - used to evaluate the performance of a classification model
#confusion_matrix - used to describe the performance of a classification model. It shows the number of correct and incorrect predictions made by the model.
```
